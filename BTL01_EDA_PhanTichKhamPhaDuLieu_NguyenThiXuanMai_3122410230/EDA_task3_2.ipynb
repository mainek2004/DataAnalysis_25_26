#Import thư viện

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
penguins_data = pd.read_csv("/content/drive/MyDrive/DataAnalysis/Tuần 5/penguins_size.csv")
penguins_data = penguins_data[['species','culmen_length_mm','body_mass_g']]
```

#Phương pháp Scatterplot

## Case 1

```python
sns.scatterplot(x='culmen_length_mm', y='body_mass_g', data=penguins_data)
```

## Case 2

```python
plt.figure(figsize=(12, 6))
ax = sns.scatterplot(x='culmen_length_mm', y='body_mass_g', data=penguins_data)
ax.set_title('Bivariate analysis of Culmen Length and Body Mass', fontsize=20)
```

# Phân tích 2 biến dựa vào bảng crosstab/two-way

```python
pd.crosstab(penguins_data['species'], penguins_data['culmen_length_mm'])
```

#Phân tích 2 biến sử dụng pivot_table

```python
pd.pivot_table(penguins_data, values='culmen_length_mm', index='species', aggfunc= np.mean)
```

#Phân tích 2 biến sử dụng pairplot

```python
sns.pairplot(penguins_data)
```
