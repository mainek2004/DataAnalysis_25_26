```python
import numpy as np
from scipy import stats

# B·ªè NaN ƒë·ªÉ t√≠nh to√°n
new_cases = covid_data["new_cases"].dropna()

data_mean = np.mean(new_cases)
data_median = np.median(new_cases)
data_mode = stats.mode(new_cases, keepdims=True).mode[0]
data_variance = np.var(new_cases)
data_sd = np.std(new_cases)
data_range = new_cases.max() - new_cases.min()
data_percentile = np.percentile(new_cases, 60)
q1 = np.percentile(new_cases, 25)
q2 = np.percentile(new_cases, 50)
q3 = np.percentile(new_cases, 75)
data_IQR = stats.iqr(new_cases)

print("Mean:", data_mean)
print("Median:", data_median)
print("Mode:", data_mode)
print("Variance:", data_variance)
print("Standard Deviation:", data_sd)
print("Range:", data_range)
print("60th Percentile:", data_percentile)
print("Q1:", q1, "Q2:", q2, "Q3:", q3)
print("IQR:", data_IQR)
```

```python
# C√†i ƒë·∫∑t th∆∞ vi·ªán
!pip install ipywidgets plotly --quiet
from google.colab import output
output.enable_custom_widget_manager()

import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display, clear_output
import time

# ƒê·ªçc v√† x·ª≠ l√Ω d·ªØ li·ªáu
path = "/content/drive/MyDrive/PTDL/bt tuan4/covid-data.csv"
df = pd.read_csv(path)

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])
df['year'] = df['date'].dt.year
df = df[(df['year'] >= 2018) & (df['year'] <= 2024)]

# Trung b√¨nh 7 ng√†y
df = df.sort_values(['location','date'])
df['new_cases_7d_avg'] = df.groupby('location')['new_cases'].rolling(7, min_periods=1).mean().reset_index(0, drop=True)

# Gom theo nƒÉm
df_yearly = df.sort_values('date').groupby(['iso_code','location','year'], as_index=False).last()

# TƒÉng tr∆∞·ªüng %
df['cases_7d_sum'] = df.groupby('location')['new_cases'].rolling(7).sum().reset_index(0, drop=True)
df['cases_prev7d_sum'] = df.groupby('location')['new_cases'].shift(7).rolling(7).sum().reset_index(0, drop=True)
df['growth_rate_cases'] = (df['cases_7d_sum'] - df['cases_prev7d_sum']) / df['cases_prev7d_sum'] * 100
df_growth_yearly = df.sort_values('date').groupby(['iso_code','location','year'], as_index=False).last()

# ====================================
# 3Ô∏è‚É£ H√ÄM V·∫º BI·ªÇU ƒê·ªí
# ====================================
def plot_map(year):
    df_map = df_yearly[df_yearly['year'] == year]
    fig = px.choropleth(
        df_map,
        locations="iso_code",
        color="total_cases_per_million",
        hover_name="location",
        color_continuous_scale="YlOrRd",
        range_color=(0, df_yearly['total_cases_per_million'].max()),
        title=f"üåç B·∫£n ƒë·ªì COVID-19 tr√™n 1 tri·ªáu d√¢n ({year})",
        projection="natural earth"
    )
    fig.show()

def plot_line(countries):
    df_line = df[df['location'].isin(countries)]
    fig = px.line(
        df_line,
        x="date",
        y="new_cases_7d_avg",
        color="location",
        title=f"üìà Trung b√¨nh 7 ng√†y s·ªë ca COVID-19 ({', '.join(countries)})"
    )
    fig.update_traces(mode="lines+markers")
    fig.show()

def plot_bar(year):
    df_bar = df_yearly[df_yearly['year'] == year].nlargest(10, "total_cases_per_million")
    fig = px.bar(
        df_bar,
        x="location",
        y="total_cases_per_million",
        color="total_cases_per_million",
        color_continuous_scale="Reds",
        text="total_cases_per_million",
        title=f"üèÜ Top 10 qu·ªëc gia c√≥ s·ªë ca COVID-19 cao nh·∫•t ({year})"
    )
    fig.show()

def plot_growth(year):
    df_g = df_growth_yearly[df_growth_yearly['year'] == year]
    fig = px.choropleth(
        df_g,
        locations="iso_code",
        color="growth_rate_cases",
        hover_name="location",
        color_continuous_scale="RdBu_r",
        range_color=(-100, 100),
        title=f"üìä T·ªëc ƒë·ªô tƒÉng tr∆∞·ªüng ca COVID-19 (%) ({year})",
        projection="natural earth"
    )
    fig.show()

# ====================================
# 4Ô∏è‚É£ GIAO DI·ªÜN T∆Ø∆†NG T√ÅC + AUTO-PLAY
# ====================================
chart_selector = widgets.ToggleButtons(
    options=[('üåç Map', 'map'), ('üìà Line', 'line'),
             ('üèÜ Bar', 'bar'), ('üìä Growth', 'growth')],
    description='Ch·ªçn bi·ªÉu ƒë·ªì:'
)

country_dropdown = widgets.SelectMultiple(
    options=sorted(df['location'].unique()),
    value=["Vietnam", "India"],
    description='Qu·ªëc gia:',
    layout=widgets.Layout(width='50%', height='180px')
)

year_slider = widgets.IntSlider(
    value=2020, min=2018, max=2024, step=1,
    description='NƒÉm:', continuous_update=False
)

output = widgets.Output()

# N√∫t l√†m m·ªõi & n√∫t ph√°t
refresh_button = widgets.Button(
    description='üîÑ Hi·ªÉn th·ªã bi·ªÉu ƒë·ªì',
    button_style='success',
    icon='refresh'
)

play_button = widgets.ToggleButton(
    value=False,
    description='‚ñ∂Ô∏è Ph√°t Map',
    button_style='info'
)

# --- H√†m c·∫≠p nh·∫≠t ---
def update_dashboard(*args):
    with output:
        clear_output(wait=True)
        chart_type = chart_selector.value
        year = year_slider.value
        countries = country_dropdown.value

        if chart_type == 'map':
            plot_map(year)
        elif chart_type == 'line':
            plot_line(countries)
        elif chart_type == 'bar':
            plot_bar(year)
        elif chart_type == 'growth':
            plot_growth(year)

# --- Auto play map ---
def autoplay_map(change):
    if play_button.value:
        play_button.description = "‚è∏ D·ª´ng"
        for y in range(2018, 2025):
            if not play_button.value:
                break
            year_slider.value = y
            update_dashboard()
            time.sleep(1.2)  # t·ªëc ƒë·ªô chuy·ªÉn nƒÉm
        play_button.value = False
        play_button.description = "‚ñ∂Ô∏è Ph√°t Map"
    else:
        play_button.description = "‚ñ∂Ô∏è Ph√°t Map"

# --- G√°n s·ª± ki·ªán ---
chart_selector.observe(update_dashboard, names='value')
year_slider.observe(update_dashboard, names='value')
country_dropdown.observe(update_dashboard, names='value')
refresh_button.on_click(update_dashboard)
play_button.observe(autoplay_map, names='value')

# --- Hi·ªÉn th·ªã dashboard ---
display(chart_selector, year_slider, country_dropdown,
        widgets.HBox([refresh_button, play_button]), output)

update_dashboard()
```

 B∆∞·ªõc 3: Th·ªëng k√™ chi ti·∫øt v·ªõi numpy & scipy

```python
import numpy as np
from scipy import stats
import pandas as pd

# ƒê·ªçc d·ªØ li·ªáu diabetes.csv
diabetes_data = pd.read_csv("/content/drive/MyDrive/PTDL/bt tuan4/diabetes.csv")

# L·∫•y danh s√°ch c√°c c·ªôt s·ªë
numeric_cols = diabetes_data.select_dtypes(include=[np.number]).columns.tolist()
results = {}

# T√≠nh to√°n th·ªëng k√™ chi ti·∫øt cho t·ª´ng c·ªôt s·ªë
for col in numeric_cols:
    data = diabetes_data[col].dropna()
    results[col] = {
        "Mean": np.mean(data),
        "Median": np.median(data),
        "Mode": stats.mode(data, keepdims=True).mode[0],
        "Variance": np.var(data),
        "Std Dev": np.std(data),
        "Range": np.max(data) - np.min(data),
        "Q1": np.percentile(data, 25),
        "Q2 (Median)": np.percentile(data, 50),
        "Q3": np.percentile(data, 75),
        "IQR": stats.iqr(data)
    }

# Xu·∫•t k·∫øt qu·∫£ ra DataFrame
desc_stats = pd.DataFrame(results).T
print("\n Th·ªëng k√™ chi ti·∫øt cho c√°c bi·∫øn s·ªë:")
print(desc_stats)
```

B∆∞·ªõc 5: Thay th·∫ø trung v·ªã cho c√°c gi√° tr·ªã 0 kh√¥ng h·ª£p l·ªá

```python
import pandas as pd
import numpy as np

# ƒê·ªçc d·ªØ li·ªáu
df = pd.read_csv("/content/drive/MyDrive/PTDL/bt tuan4/diabetes.csv")

# C√°c c·ªôt kh√¥ng th·ªÉ c√≥ gi√° tr·ªã 0 v·ªÅ m·∫∑t y t·∫ø
invalid_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

# Thay th·∫ø 0 b·∫±ng median c·ªßa t·ª´ng c·ªôt
for col in invalid_cols:
    median_value = df[col].median()
    df[col] = df[col].replace(0, median_value)

# Ki·ªÉm tra l·∫°i xem c√≤n gi√° tr·ªã 0 kh√¥ng
print((df[invalid_cols] == 0).sum())

# L∆∞u file m·ªõi sau khi x·ª≠ l√Ω
df.to_csv("/content/drive/MyDrive/PTDL/bt tuan4/diabetes_new.csv", index=False)
print("‚úÖ ƒê√£ thay th·∫ø xong gi√° tr·ªã 0 b·∫±ng median, file m·ªõi: diabetes_new.csv")
```

B∆∞·ªõc 6: Histogram so s√°nh tr∆∞·ªõc v√† sau khi x·ª≠ l√Ω

```python
import pandas as pd
import matplotlib.pyplot as plt

# ƒê·ªçc d·ªØ li·ªáu g·ªëc
df_raw = pd.read_csv("/content/drive/MyDrive/PTDL/bt tuan4/diabetes.csv")

# T·∫°o b·∫£n sao v√† thay th·∫ø gi√° tr·ªã 0 b·∫±ng median
df_clean = df_raw.copy()
invalid_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for col in invalid_cols:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].replace(0, median_value)

# V·∫Ω histogram so s√°nh tr∆∞·ªõc & sau x·ª≠ l√Ω
for col in invalid_cols:
    plt.figure(figsize=(12,5))

    # Tr∆∞·ªõc x·ª≠ l√Ω
    plt.subplot(1,2,1)
    plt.hist(df_raw[col], bins=30, color="salmon", edgecolor="black")
    plt.title(f"{col} - Tr∆∞·ªõc x·ª≠ l√Ω")
    plt.xlabel(col)
    plt.ylabel("T·∫ßn su·∫•t")

    # Sau x·ª≠ l√Ω
    plt.subplot(1,2,2)
    plt.hist(df_clean[col], bins=30, color="skyblue", edgecolor="black")
    plt.title(f"{col} - Sau x·ª≠ l√Ω (thay b·∫±ng median)")
    plt.xlabel(col)
    plt.ylabel("T·∫ßn su·∫•t")

    plt.tight_layout()
    plt.show()
```

B∆∞·ªõc 6: V·∫Ω Boxplot (ph√°t hi·ªán outliers- gi√° tr·ªã ngo·∫°i lai; gi√° tr·ªã r·∫•t l∆°n ho·∫∑c r·∫•t nh·ªè so v·ªõi ph√¢n ph·ªëi t·ªïng th·ªÉ, c√≥ th·ªÉ l√†m ·∫£nh h∆∞·ªüng ƒë·∫øn t√≠nh ch√≠nh x√°c c·ªßa m√¥ h√¨nh)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ƒê·ªçc d·ªØ li·ªáu g·ªëc
df_raw = pd.read_csv("/content/drive/MyDrive/PTDL/bt tuan4/diabetes.csv")

# T·∫°o b·∫£n sao v√† x·ª≠ l√Ω 0 b·∫±ng median
df_clean = df_raw.copy()
invalid_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]

for col in invalid_cols:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].replace(0, median_value)

# L·∫•y danh s√°ch c√°c c·ªôt s·ªë
numeric_cols = df_raw.select_dtypes(include=["number"]).columns.tolist()

# Boxplot tr∆∞·ªõc x·ª≠ l√Ω
plt.figure(figsize=(15,8))
sns.boxplot(data=df_raw[numeric_cols])
plt.title("Boxplot tr∆∞·ªõc x·ª≠ l√Ω (c√≤n nhi·ªÅu gi√° tr·ªã 0 v√† outlier)")
plt.xticks(rotation=45)
plt.show()

# Boxplot sau x·ª≠ l√Ω
plt.figure(figsize=(15,8))
sns.boxplot(data=df_clean[numeric_cols])
plt.title("Boxplot sau x·ª≠ l√Ω (ƒë√£ thay 0 b·∫±ng median)")
plt.xticks(rotation=45)
plt.show()
```
