```python
import pandas as pd
from autoviz import AutoViz_Class

# 1) Đọc TSV
file_path = "/content/drive/MyDrive/DataAnalysis/Tuần 5/marketing_campaign.csv"
df = pd.read_csv(file_path, sep="\t")

# 2) Làm sạch gọn
df = df.drop_duplicates()
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], errors="coerce", dayfirst=True)

# Gộp rare categories
df["Marital_Status"] = df["Marital_Status"].replace({"Alone":"Other","Absurd":"Other","YOLO":"Other"})

# Điền thiếu Income
if df["Income"].isna().any():
    df["Income"] = df["Income"].fillna(df["Income"].median())


AV = AutoViz_Class()
_ = AV.AutoViz(
    filename="",
    dfte=df,
    depVar="Response",
    chart_format="svg",
    verbose=2,
    max_rows_analyzed=200000,
    max_cols_analyzed=100,
    save_plot_dir="autoviz_reports"
)
print("Charts saved in autoviz_reports/")
```
